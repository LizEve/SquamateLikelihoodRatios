## Problem

**Intermediate:**  >sqc-3884 SRP052050_contig113

**Final:**  >AHE-L383 RP052050_contig113

## Approach

1. Pull out sqc and scaffold # 
    - a. regex
    - b. in/out file
2. Use probes_sqc.fasta to rename sqc with ahe/uce/gene
    - a. read in sqc and prob name, make dictionary?
    - b. match sqcs from genome.fasta and probe.fasta 
3. Replace record id in genome file with original probe name

```python
#2.a Read in sqc and prob name, make dictionary?
# 37517 total probes match what is reported in the singhal paper. 5358 loci

import re
from Bio import SeqIO

probe_file = "probes_sqc.fasta"
sqcProbeDict={}
sqcList=[]

for record in SeqIO.parse(probe_file, 'fasta'):
    d=re.split("\|",record.description)
    ##['sqc-1_p4 ', 'source: Anolis-uce-1869-p1-UCE_60']
    sqc = re.split("_",d[0])[0]
    sqcList.append(sqc)
    longName = re.split("_|-",d[1])
    ##['source: Gekko', 'uce', '3922', 'p1', 'UCE', '73']
    ##['source: anoCar2', 'L118', 'AHE', '0']
    ##['source: Gekko', 'BACH1', 'genes', '126']
    if longName[2] == "AHE":
        newName=longName[2]+"-"+str(longName[1])   
    elif longName[2] == "genes":    
        newName="gene-"+str(longName[1]) 
    elif longName[1] == "uce":
        newName=longName[1]+"-"+str(longName[2])  
    sqcProbeDict[sqc]=newName
print(len(sqcProbeDict))
```

```python
#1.b and 3 match sqcs from genome.fasta and probe.fasta 
import re
from Bio import SeqIO

original_file = "ophigra.fasta"
corrected_file = "Ophiosaurus_gracilis.fasta"
with open(corrected_file, 'w') as corrected:
    for key, value in sqcProbeDict.iteritems():
        for record in SeqIO.parse(original_file, 'fasta'):
            # save original record id before altering it 
            oldHeader = record.id
            n = re.split("\|",oldHeader)
            sqcName = re.split(":",n[3])[1]
            contig = "RP052050_contig"+re.split("d",re.split(":",n[1])[1])[1]
            if sqcName == key:
                record.id = value
                record.description = contig
                print(sqcName)
                print(record.id)
                SeqIO.write(record, corrected, 'fasta')
```

```python
import re
import os
from Bio import SeqIO

def renameFasta(original_file,corrected_file,ident,probe_file):
    sqcProbeDict={}
    sqcList=[]
    for record in SeqIO.parse(probe_file, 'fasta'):
        d=re.split("\|",record.description)
        ##['sqc-1_p4 ', 'source: Anolis-uce-1869-p1-UCE_60']
        sqc = re.split("_",d[0])[0]
        sqcList.append(sqc)
        longName = re.split("_|-",d[1])
        ##['source: Gekko', 'uce', '3922', 'p1', 'UCE', '73']
        ##['source: anoCar2', 'L118', 'AHE', '0']
        ##['source: Gekko', 'BACH1', 'genes', '126']
        if longName[2] == "AHE":
            newName=longName[2]+"-"+str(longName[1])
        elif longName[2] == "genes":
            newName="gene-"+str(longName[1])
        elif longName[1] == "uce":
            newName=longName[1]+"-"+str(longName[2])
        sqcProbeDict[sqc]=newName
    print(len(sqcProbeDict))

    with open(corrected_file, 'w') as corrected:
        for key, value in sqcProbeDict.iteritems():
            for record in SeqIO.parse(original_file, 'fasta'):
                # save original record id before altering it
                oldHeader = record.id
                n = re.split("\|",oldHeader)
                sqcName = re.split(":",n[3])[1]
                contig = str(ident)+"_contig"+re.split(":",n[1])[1]
                if sqcName == key:
                    record.id = value
                    record.description = contig
                    print(sqcName)
                    print(record.id)
                    SeqIO.write(record, corrected, 'fasta')

probe_file = "/Users/ChatNoir/Projects/Squam/scripts/GenomeMining/probes_sqc.fasta"
mainDir = "/Users/ChatNoir/Projects/Squam/scripts/GenomeMining"
os.chdir(mainDir)
renameFasta("gekjap1.fasta","Gekko_japonicus.fasta","GekkojaponicusV1.1",probe_file)

```
